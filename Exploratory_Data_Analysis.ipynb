## **2. Creating Data Frame for Loss**
Few sales are in losses. Therefore, a separate dataframe is created to calculate total losses.

```python
sales_loss = sales[sales['Profit'] < 0]
sales_loss.head()
```

```python
# Rows and Columns of Loss Data
sales_loss.shape
```

```python
sales_loss.info()
```

```python
sales_loss.describe()
```

```python
sales_loss.describe(include='O')
```

## **Information about the Loss Sales Data:**

- **Office Supplies** has facing most losses
- **Binders** has high numbers of sales and also facing losses

Loss Sales orders occurence Location Wise:
  - **Region:** Central
  - **State:** Texas
  - **City:** Philadelphia

```python
# Calculate Total Loss
sales_loss['Profit'].sum()
```

### Total Loss is **156,131.29**

```python
sales_loss.groupby('Segment')[['Quantity', 'Sales', 'Discount' ,'Profit']].sum().sort_values('Profit', ascending =True)
```

### High Discounts lead to high losses





```python
# Grouping Categories and Subcategories and sum up total quantity, sales, discount and profit

sales_loss.groupby(['Category', 'Sub-Category'])[['Quantity', 'Sales', 'Discount' ,'Profit']].sum().sort_values(['Category', 'Profit'], ascending = [True, True])
```

**Total Losses Categories and SubCategories**
- **Tables** in Furniture Category 
- **Binders** in Office Supplies
- **Machines** in Technology

```python
# Total Sales Order to find top selling loss-making items:
sales_loss['Sub-Category'].value_counts()
```

```python
# Top 10 Loss making States
sales_loss.groupby('State')[['Quantity', 'Sales', 'Discount' ,'Profit']].sum().sort_values('Profit', ascending =True).head(10)
```

## **Visualization**

```python
# Total Sales Amount on all categories
sns.catplot(x='Sub-Category', y='Sales', data = sales, kind='bar', ci =None,  height=5, aspect=3)

#Setting labels and titles
plt.xlabel("SubCategories", fontsize= 12)
plt.ylabel("Sales Amount", fontsize= 10)
plt.title("Total Sales (Sub-Category Wise)", fontsize = 20)

plt.show()
```

### Sales is very low in **Fasteners, Art, Labels** among all sales orders

```python
# Sales Amount on Loss making Categories

sns.catplot(x='Sub-Category', y='Sales', data = sales_loss, kind='bar', ci =None,  height=5, aspect=3)

#Setting labels and titles
plt.xlabel("SubCategories", fontsize= 12)
plt.ylabel("Sales Amount", fontsize= 10)
plt.title("Total Sales (Sub-Category Wise)", fontsize = 20)

plt.show()
```

### Sales is very low in **Fasteners, Binders, Appliances\** among loss sales orders

```python
# Sales Amount on Loss making Categories

sns.catplot(x='Sub-Category', y='Discount', data = sales_loss, kind='bar', ci =None,  height=5, aspect=3)

#Setting labels and titles
plt.xlabel("SubCategories", fontsize= 12)
plt.ylabel("Discount Amount", fontsize= 10)
plt.title("Total Discounts (Sub-Category Wise)", fontsize = 20)

plt.show()
```

### Discounts is very high in **Binders and Appliances\** among loss sales orders. That's why, their sales amount is low and occur losses.

```python
# Sales Order in Ship Mode
sns.set_palette("RdBu_r")

sns.catplot(x='Ship Mode', data = sales, kind='count', ci=None,  height=4, aspect=2)

#Setting labels and titles
plt.xlabel("Ship Mode", fontsize= 12)
plt.ylabel("No. of Orders", fontsize= 10)
plt.title("Total Sales Order (Ship Mode Wise)", fontsize = 20)


plt.show()
```

### Mostly sales orders carried our in **Standard Class Shipmode**

```python
# Numbers of orders received in each segment
sns.catplot(x='Segment', data=sales, kind='count',  height=4, aspect=2)

#Setting labels and titles
plt.xlabel("Segment", fontsize= 12)
plt.ylabel("No. of Orders", fontsize= 10)
plt.title("Total Sales Order (Segment Wise)", fontsize = 20)

plt.show()
```

### Highest Sales orders received in **Consumer Segments**

```python
# Number of Sales order received in Sub-categories

plt.figure(figsize=(20,5))
sns.countplot(x='Sub-Category', data=sales) 

#Setting labels and titles
plt.xlabel("Sub-Categories", fontsize= 12)
plt.ylabel("No. of Orders", fontsize= 10)
plt.title("Total Sales Order (Sub-Categories Wise)", fontsize = 20)

plt.show()
```

### Low number of orders received in **Machines, Copiers**

```python
# Number of sales in each Region
plt.figure(figsize=(7,3))

sns.catplot(x='Region', data=sales, kind='count',  height=4, aspect=2) 

#Setting labels and titles
plt.xlabel("Region", fontsize= 12)
plt.ylabel("No. of Orders", fontsize= 10)
plt.title("Total Sales Order (Region Wise)", fontsize = 20)

plt.show()
```

```python
# Visualize correlations in Heatmap

plt.figure(figsize=(10,5))
sns.heatmap(sales.corr(), annot = True)
```

- There is a negative correlation between Sales and Discount. **(-0.22)**

- There is a moderate positive correlation between Sales and Profit **(0.48)**

## **Recommendations** 

-	Standard Shipment Mode should be used as it helps in increasing profits. 
-	States (Texas, Ohio, Pennsylvania) and cities (Philadelphia, Houston, San Antonio) are the ones where the company needs to reanalyze their strategies as they are in losses.
-	These States and cities are also the ones who are giving high amounts of discounts, so the company needs to reduce the discounts to avoid losses. 
-	Discounts on Loss making categories should be reduced to increase the profit. 
-	Discount offerings on all products should be recalculated to avoid losses.
- The Home office segment needs better improvement.
